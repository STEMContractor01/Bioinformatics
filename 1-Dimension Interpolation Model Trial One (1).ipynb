```python
#1-Dimension Interpolation Model
#Import the Packages
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
```

```python
#Defining x and y arrays of the initial data set 
x = np.linspace(0, 360, 50) 
y = (x*28)//60

```

```python
#X Array & Y Array That Will Be Used for Interpolating New Point Values
x_new = np.linspace(0, 340, 100)
```

```python
#1-Dimension Interpolation For Loop
kind = ['linear', 'nearest', 'zero', 'slinear', 'quadratic', 'cubic', 'previous', 'next']
fig = plt.figure()
ax = fig.subplots()

for i in kind:
    #interpolation step
    f = interpolate.interp1d(x, y, kind = i)
    #y array that contains the interpolated data points
    y_interp = f(x_new)
    ax.plot(x_new, y_interp, alpha=0.5, label = i)
    
ax.scatter(x,y)
plt.legend()
plt.show()
```
